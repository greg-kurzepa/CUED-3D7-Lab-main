```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os

results_folder = "results"

filenames = [
    "asymmetric_results_logarithmic.csv",
    "asymmetric_results_quadratic.csv",
    "asymmetric_results_quartic.csv",
]
labels = [
    "logarithmic",
    "quadratic",
    "quartic",
]

dfs = []
for fn in filenames:
    dfs.append(pd.read_csv(os.path.join(results_folder, fn)))

fig, ax = plt.subplots(1,2, sharey=True)

for idx, df in enumerate(dfs):
    ax[0].plot(df["Element Count"], df["Max Von Mises Stress (Mpa)"], label=labels[idx], linestyle='--', marker='o')
    ax[1].plot(df["Elapsed Time (s)"], df["Max Von Mises Stress (Mpa)"], label=labels[idx], linestyle='--', marker='o')

ax[0].grid()
ax[0].set_xlabel("Element Count")
ax[0].set_ylabel("Maximum Von Mises Stress (MPa)")
ax[0].ticklabel_format(axis='x', style='sci', scilimits=(0, 0))
ax[0].set_xlim(left=0)
ax[1].grid()
ax[1].legend()
ax[1].set_xlim(left=0)
ax[1].set_xlabel("Elapsed Time (seconds)")
plt.suptitle("Convergence of Various Mesh Element Size Distributions", size=14, y=0.95)
plt.subplots_adjust(wspace=0.05)

# plt.savefig("results\\asymmetric_experiments.png", dpi=300)
plt.show()
```
